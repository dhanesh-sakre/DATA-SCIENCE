```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ppscore as pps
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/dhanesh-sakre/DATA-SCIENCE/refs/heads/main/Feature%20Engineering/Predictive%20Power%20Score/Iris.csv', index_col=0)
df
```

I have to find out correlation between independent features and target column(Species)

pps by default creates decision tree to find out the relation between two columns and it calculates F1-SCORE by default

```python
# Identifying correlation between SepalLengthCm independent feature and  target column(Species)

pps.score(df, x='SepalLengthCm', y='Species')
```

```python
# interpretation:
# SepalLengthCm is moderately impacting the target column that means there is not a strong relationship between these two columns('SepalLengthCm' to 'Species').
```

```python
# Now I want to find out correlation between all independent features and target column(Species) for that I will use matrix() function with pps
# metrix will create all the possible combination of columns and calculates pps
```

```python
feature_selection = pps.matrix(df)
```

```python
feature_selection
```

```python
# pps between x to y  and y to x has been calculated
# it is calculating pps between same colum also
# and pps between x to x also
```

I want to find out pps between independent features and target column(Species) only

```python
# using pandas filtering to get only desired output

feature_imp = feature_selection[(feature_selection['y']=='Species') & (feature_selection['x'] != 'Species')]
feature_imp
```

```python
feature_imp.sort_values(by='ppscore', ascending=False)
```

I can see that PetalWidthCm column is having the highest impact and SepalWidthCm column  is having the least impact on the taget column according to ppscore

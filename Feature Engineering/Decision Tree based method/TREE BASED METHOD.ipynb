# defining x and y varibles

<p  style = "font-size : 35px; color : #34656d ; font-family : 'Comic Sans MS'; text-align : center; background-color : #00008B; border-radius: 5px 5px;"><strong>TREE BASED METHOD</strong></p> 

```python
# I need to build decision tree in order to find out information gain for each indepedent features
```

```python
# I need to pass whole data in order to know which features are important and which one not
# it will calculate entropy and information gain for all the columns

dt = DecisionTreeClassifier(criterion='entropy')
dt.fit(x, y)
```

```python
# asking decision tree which columns are important and highly related to target columns by using feature_importances_ function

dt.feature_importances_
```

```python
# but this is not easily intrepretable so converting it into dataframe datastructure
ig_sel = pd.DataFrame({
    'features' : x.columns,
    'info_gain' : dt.feature_importances_
})
```

```python
# sorting the columns by information gain so that I can easily find out which columns are the most important one

ig_sel.sort_values(by='info_gain', ascending=False)
```

```python
# importing liberaries:
# for data maupulation
import pandas as pd
import numpy as np
# for ignoring future warnings
import warnings
warnings.filterwarnings('ignore')

# for NLP
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import re
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

# DATA CLEANING

```python
# I have to remove all the emojis, special characters etc..
# because I only want letters that are present in the upper and lower case
```

```python
# using a loop and regular expression liberary to extract words
# and storing them in text variable
# then converting them in lower case
# then appended them in the corpus list

corpus = []

for i in range(len(sent)):
  text = re.sub('[^a-zA-Z]', ' ', sent[i])
  text = text.lower()
  corpus.append(text)
```

```python
corpus
```

```python
# all the extra part removed except words
# only clean sentences are present
```

```python
# NOW I need to remove topword and also need to apply stemming
```

```python
# creating a instance for porterstemmer

stemmer = PorterStemmer()
```

```python
# bro let's see how it works
# appying stemming on argued

stemmer.stem('argued')
```

```python
# although stemming is not that useful but I need to know about it
# stemming is not an effiecient method because it is not able to capture original words properly
# I will always or most of the time use lemmatization
```

```python
# now download the package for removing stopwords

nltk.download('stopwords')
```

```python
# let's see the stopwords present in the english language

stopwords.words('english')
```

```python
# the stopwords present in the French language
stopwords.words('french')
```

```python
# Now I need to convert the sentences present in the corpus list to words(word tokenization)
# let's extract all the words that are present in the corpos

for i in corpus:
  words = word_tokenize(i)
  for i in words:
    print(i)
```

```python
# all the words that are present in the corpus
```

# STEMMING

```python
# now applying stemming on this words
```

```python
# if the word is not a stop word then I will store print it
for i in corpus:
  words = word_tokenize(i)
  for i in words:
    if i not in stopwords.words('english'):
      print(stemmer.stem(i))
```

```python
# stop words removed
```

# LEMMATIZATION

```python
# creating a object for the class
lemma = WordNetLemmatizer()

# installing dependencies
nltk.download('wordnet')
```

```python
for i in corpus:
  words = word_tokenize(i)
  for i in words:
    if i not in stopwords.words('english'):
      print(lemma.lemmatize(i))
```

```python
# I can compare the results of lemmatization and stemming
# I can  see that lemmatization is giving me better results
```

# Feature Extraction

```python
# there are 3 methods to do it:
# 1. Bag of words
# 2. Binary weight
# 3. TF-IDF
```



# 1. BAG OF WORDS (BoW)



```python
# I need to use countvectorizor to imnplement bag of words method.

cv = CountVectorizer()
x = cv.fit_transform(corpus)
x
```

```python
# it will see how many unique words are presentin the corpus
# their frequency
# transform will put them into the tabular format(in 1, 0 format)
```

```python
# convert it to array
x.toarray()
```

```python
# CONVERTING IT INTO DATAFRAME
pd.DataFrame(x.toarray())
```

```python
# here on the columns name I want unique word instead of index number
```

```python
# see the vocabulary containe in this dictonary
cv.vocabulary
```

```python
# getting a list of the unique words extracted from corpus by Bag of words

cv.get_feature_names_out()
```

```python
# using these unique words in the column
pd.DataFrame(x.toarray(), columns=cv.get_feature_names_out())
```

```python
# row are presenting the 8 sentences
# and unique words are present as the column name
# 1 is indicating the presence of the word in that sentence
# 0 is indicating the absence of the word in that sentence
```

# BINARY WEIGHT

```python
# same class is used for both:
# bag of words(BoW) and Binary Weight
# only difference is the parameter in it "binary = True" for binary weight


bw = CountVectorizer(binary=True)
x = bw.fit_transform(corpus)
pd.DataFrame(x.toarray(), columns=bw.get_feature_names_out())
```

# TF-IDF

```python
# creating an instance for class
tf = TfidfVectorizer()

# fitting the cleaned text
data = tf.fit_transform(corpus)
data
```

```python
# now converting it into array format
data.toarray()
```

```python
# it has returned weight value for each unique word in the tabular format instead of 0 and 1 unlike BoW and binary weight
```

```python
# converting to dataframe so that it can be interpretable
x = pd.DataFrame(data.toarray(), columns=tf.get_feature_names_out())
x
```

```python
# the purpose of these 3 technique is same but the only difference is how they fill the table
```

```python
# binayr weight ---> 0, 1
# bag of words ----> frequency of word
# TF-IDF     ------> weight value of that word
```

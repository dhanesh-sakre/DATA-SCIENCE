# Conversion of Categorical Data into Dummy Variables

```python
categorical_variables = df.iloc[:,[0,1,2,3,5,7]]
categorical_variables
```

```python
categorical_variables.nunique()
```

```python
categorical_variables.nunique().sum()
```

```python
# there will be total 86 new columns after applying ONE HOT ENCODING
```

```python
dummy_variables = pd.get_dummies(categorical_variables, dtype = int)
```

here 0 represent that the value is not present in that newly created column
and 1 represent that the value is present in that newly created column

### NOW THE DATA IS TRANSFORMED

```python
dummy_variables
```

```python
# now there ,450 rows and 86 columns are present after applying ONE HOT ENCODING
```

# defining the function for outlier removal by using capping method

```python
def capping(data, columns) :
    for col in columns:
        q1 = data[col].quantile(0.25)
        q3 = data[col].quantile(0.75)
        IQR = q1 - q3
        lower = q1 - (1.5 * IQR)
        upper = q3 + (1.5 * IQR)
        df[col]  = np.where( df[col] > upper, upper,
                            np.where(df[col] < lower, lower,
                                     df[col])
                           )
```

calling the functiion to remove outliers from all the columns

```python
capping(df, df.columns)
```

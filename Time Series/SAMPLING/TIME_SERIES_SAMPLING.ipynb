```python
# reading data and converted date column into index column
# telling python that this is a data related to time series

df = pd.read_csv('https://raw.githubusercontent.com/dhanesh-sakre/DATA-SCIENCE/refs/heads/main/Time%20Series/SAMPLING/sales.csv', index_col=0, parse_dates=True)
df
```

# UPSAMPLING 

```python
# here I have performed upsamplling and increased the frequency of the data
```

```python
# this is an example of upsampling like how to perform upsampling
Upsampled = df.resample('D').mean()
Upsampled.head(50)
```

# INTERPOLATING these missing values 

```python
interpolated = Upsampled.interpolate()
interpolated.head(50)
```

```python
# upsamplled dataset
interpolated.shape
```

```python
# original dataset
df.shape
```

```python
# see the difference between the original dataset and upsamplled dataset
# I can see that in the new dataset there are 1066 datapoints present whereas in the original dataset there are 36 datapoints present
```

```python
# plotting line plot for interpolatted  data
interpolated.plot()
```

```python
# observe:

# although I changed the number of datapoints but the distribution of the data is same
```

# Downsamplling

```python
# PERFORMING DOWNSAMPLING TO GET QUARTERLY SALES DATA

downsampling = df.resample('Q').mean()
downsampling
```

```python
# plotting graph on downsamplled data
downsampling.plot()
```

```python
# although datapoints decreased but the trend is same as it was earlier
```

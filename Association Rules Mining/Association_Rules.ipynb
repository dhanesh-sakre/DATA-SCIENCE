```python
# importing necessary liberaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as wr
wr.filterwarnings('ignore')

# from mlxted liberay importing modules and fuctions
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns  import apriori, association_rules
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/aishwaryamate/Datasets/refs/heads/main/Supermarket.csv')
df
```

# DATA TRANSFORMATION

```python
# converting each transaction into multiple values of products contained by that perticular transaction

data = [i.split(',') for i in df['Products']]
data
```

```python
# Encoding text values of each products for further use

te = TransactionEncoder()
encoded_df = te.fit_transform(data)
encoded_df
```

```python
# unique products exctrated by TransactionEncoder class

te.columns_
```

```python
data_df = pd.DataFrame(encoded_df, index=df['ID'], columns=te.columns_)
data_df
```

```python
# TO understand the data I am converting it into dataframe
```

```python
data_df.replace([True, False], [1, 0], inplace=True)
data_df
```

# MODEL BUILDING

```python
# using apriori algorithm
```

```python
scores = apriori(data_df, min_support=0.2, use_colnames=True)
scores
```

```python
# algorithm has calculated one-item set list, two-item set list as well as three item set list along with their support value
# and it has selected only those rules that are having support more than or equal to (>=) min_support value (0.2)
```

```python
# calculate all the matrics(support, confidence, lift ratio)
association_rules(scores, num_itemsets=2)
```

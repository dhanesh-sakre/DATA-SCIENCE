```python
# imorting necessary liberaries
# for data manupulation
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# to ignore future warnings
import warnings
warnings.filterwarnings('ignore')

# for model and prediction
import spacy
from spacy import displacy
```

```python
# loading the model
nlp = spacy.load('en_core_web_lg')
```

```python
# just to see how the NER works I am using one sentence

doc = nlp("mr. Elon Musk acquired Twitter Organization for $45 billion in 2022")
doc
```

```python
# to identify whether NER has identified named entities
doc.ents
```

```python
# in order to tag the entities I will use for loop

for i in doc.ents:
  print(i, ':', i.label_)
```

```python
# it is able to recognize the category of the entities
```

```python
# let's join these piece of text with space to get one piece of text(paragraph)
txt = ' '.join(df['x'])
txt
```

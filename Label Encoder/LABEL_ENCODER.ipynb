```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/dhanesh-sakre/DATA-SCIENCE/refs/heads/main/PCA%20(dimensionality%20reduction)/Iris.csv')
df
```

# LABEL ENCODER

```python
# defining x and y variable
x = df.drop(columns=['Species'])
y = df['Species']
```

```python
x
```

```python
y
```

```python
# calling class and created variable for it
le = LabelEncoder()

# fitting the target column because I only want to encode target column now
y_encoded = le.fit_transform(y)
y_encoded
```

```python
# for understanding in a better way I am converting it into dataframe
y_encoded = pd.DataFrame(y_encoded)
y_encoded
```

```python
# concatinating actual original target column and encoded target column
df_new = pd.concat([y, y_encoded], axis=1)
df_new.tail(20)
```

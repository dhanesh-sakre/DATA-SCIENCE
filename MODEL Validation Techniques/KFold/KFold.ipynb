2. KFold

# defining x and y varibles

```python
x = df.drop(columns=['fetal_health'])
x
```

```python
y = df['fetal_health']
y
```

<p  style = "font-size : 35px; color : #34656d ; font-family : 'Comic Sans MS'; text-align : center; background-color : #00008B; border-radius: 5px 5px;"><strong>KFOLD CROSS VALIDATION TECHNIQUE</strong></p>

```python
# call the KFold class and define how many splits of data you want to do 
# it will select random 80% datapoints for training and other 20% of datapoints for testing purpose
# KFold only decides how many folds it has to do but I need to use cross_val_score class.

fold = KFold(n_splits=5)
```

```python
# I will use Decision tree model
# cross_val_score is responsible to create model and calculate the accuracy of the model on the data that I will mention in it
# mention estimatior(model) which I will use to calculate the accuracy
# pass entire data(all independent features and target column)
# cross validation technique(method)

score = cross_val_score(DecisionTreeClassifier(),x, y, cv = fold )
```

it has created 5 models and trained those models on each fold of data the calculated the accuracy of the model on the selected datapoints

```python
# checking accuracy of all the 5 models
score
```

```python
# I have to calculate average of the accuracies to get final accuracy
score.mean()
```

this is the final accuracy on this dataset by using DecisionTreeClassifier model

# using Logistic Regression

```python
# creating 5 folds
kfol = KFold(n_splits=5)
```

```python
# calculating accuracies and storing them in scores variable

scores = cross_val_score(LogisticRegression(), x, y, cv=kfol)
```

```python
# asking for accuracy
scores
```

```python
# final accuracy of logistic regression model on this dataset
scores.mean()
```

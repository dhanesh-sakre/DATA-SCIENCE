```python
# basic liberaries for data manupulation
import pandas as pd
import numpy as np

# liberaries for ploting graphs 
import seaborn as sns
import matplotlib.pyplot as plt
# parameters of graps
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['figure.dpi'] = 250
sns.set_theme(style='darkgrid')
 
# liberaries for using cross validation methods 
from sklearn.model_selection import train_test_split, KFold, LeaveOneOut, cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import confusion_matrix, classification_report

# to ignore future warnings
import warnings
warnings.filterwarnings('ignore')
```

# defining x and y varibles

```python
x = df.drop(columns=['fetal_health'])
x
```

```python
y = df['fetal_health']
y
```

<p  style = "font-size : 35px; color : #34656d ; font-family : 'Comic Sans MS'; text-align : center; background-color : #00008B; border-radius: 5px 5px;"><strong>TRAIN TEST SPLIT</strong></p> 

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, train_size=0.20, random_state=42)
```

```python
xtrain.head()
```

```python
xtest.head()
```

```python
ytrain.head()
```

```python
ytest.head()
```

# MODEL

```python
# decision tree
dt = DecisionTreeClassifier()

# fitting the model on training data(model trainning)
dt.fit(xtrain, ytrain)
```

```python
# getting prediction
ypred = dt.predict(xtest)
ypred
```

# MODEL EVALUATION

```python
#  precision,recall,f1-score,support,accuracy

print(classification_report(ytest, ypred))
```
